```python
n = 10

results = dict()
for c in range(1, n+1):
    for d in range(1, n+1):
        result = c**3 + d**3
        try:
            results[result].append((c, d))
        except:
            results[result] = list()
            results[result].append((c, d))

for a in range(1, n+1):
    for b in range(1, n+1):
        result = (a**3) + (b**3)
        temp = results[result]
        for pair in temp:
            print(a, b, pair[0], pair[1])
        
```
